```python
df['month'] = pd.to_datetime(df['month'], format='%m/%d/%Y %I:%M:%S %p')

df_2018 = df[df['month'].dt.year == 2018]

min_visitors = df_2018.iloc[:, 1:].min()
max_visitors = df_2018.iloc[:, 1:].max()
result = pd.DataFrame({'min': min_visitors, 'max': max_visitors})

result
```

```python
average_visitors = df.iloc[:, 1:].mean()
average_visitors
```

```python
# Фільтруємо лише за 2015 рік
df_2015 = df[df['month'].dt.year == 2015].copy()

# Додаємо новий стовпець із загальною кількістю відвідувачів по всіх музеях
df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)

# Знаходимо рядок з мінімальною та максимальною кількістю
min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

# Виводимо назви місяців
min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')

print(f'Найменша відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
df_2015.head()
```

```python
df_2018 = df[df['month'].dt.year == 2018].copy()

# Витягаємо номер місяця
df_2018['month_num'] = df_2018['month'].dt.month

# Літні місяці: червень (6), липень (7), серпень (8)
summer = df_2018[df_2018['month_num'].isin([6, 7, 8])]['avila_adobe'].sum()

# Зимові місяці: грудень (12), січень (1), лютий (2)
# Потрібно об'єднати грудень 2017 та січень-лютий 2018 — але якщо лише 2018, то:
winter = df_2018[df_2018['month_num'].isin([1, 2, 12])]['avila_adobe'].sum()

print(f'Літом відвідали: {summer} людей')
print(f'Зимою відвідали: {winter} людей')
```

```python
mtx = df.corr(numeric_only=True).abs()

fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(data=df, x='month', y='avila_adobe')
plt.title('Avila Adobe: Відвідуваність з часом')
plt.show()
```

```python
df['year'] = df['month'].dt.year
sns.boxplot(data=df, x='year', y='chinese_american_museum')
plt.title('Chinese American Museum: Розподіл по роках')
plt.show()
```

```python
sns.histplot(df['gateway_to_nature_center'], bins=30, kde=True)
plt.title('Розподіл відвідувань Gateway to Nature Center')
plt.show()
```

```python
avg_visitors = df.drop(columns=['month', 'year']).mean().sort_values(ascending=False)
sns.barplot(x=avg_visitors.values, y=avg_visitors.index)
plt.title('Середня кількість відвідувачів по музеях')
plt.xlabel('Середнє значення')
plt.ylabel('Музеї')
plt.show()
```

Побудувати 5 графіків з використанням бібліотеки seaborn (не повторювати ті, що були в лекції)

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='body_mass_g', hue='species')
plt.title('Залежність довжини дзьоба від маси тіла')
plt.show()
```

```python
sns.violinplot(data=penguins, x='island', y='flipper_length_mm', hue='species', split=True)
plt.title('Розподіл довжини плавників по островах')
plt.show()
```

```python
sns.swarmplot(data=penguins, x='species', y='body_mass_g')
plt.title('Маса тіла пінгвінів по видах')
plt.show()
```

```python
corr = penguins.drop(columns=['species','island','sex']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Кореляція числових ознак')
plt.show()
```

```python
sns.catplot(data=penguins, x="island", y="flipper_length_mm", hue="sex", col="species", kind="box")
plt.suptitle("Довжина плавників по островах та статі", y=1.05)
plt.show()
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
plt.suptitle("Багатовимірні залежності ознак пінгвінів", y=1.02)
plt.show()
```
