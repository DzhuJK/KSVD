```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("iamsouravbanerjee/customer-shopping-trends-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(os.path.join(path, "shopping_trends.csv"))
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["age"], bins=20)
plt.title("Розподіл віку клієнтів")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df["item_purchased"], order=df["item_purchased"].value_counts().index)
plt.title("Найпопулярніші товари")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)', alpha=0.5)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупки')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["review_rating"], bins=10, kde=True)
plt.title("Розподіл рейтингу товарів")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gender')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
for i, v in enumerate(df['gender'].value_counts().values):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='review_rating', y='purchase_amount_(usd)', data=df, hue='gender', alpha=0.7)
plt.title("Залежність між рейтингом і сумою покупок")
plt.xlabel("Рейтинг")
plt.ylabel("Сума покупки")
plt.show()
```

```python
import altair as alt
from vega_datasets import data

wheat = data.wheat()
```

```python
chart_wheat = alt.Chart(wheat).mark_bar().encode(
    x='wheat:Q',
    y='year:O'
) + alt.Chart(wheat).mark_text(align='left', baseline='middle', dx=3).encode(
    x='wheat:Q',
    y='year:O',
    text='wheat:Q'
)

chart_wheat.properties(title="Врожайність пшениці за роками")
```

```python
df = data.iowa_electricity()

chart = alt.Chart(df).mark_bar(opacity=0.7).encode(
    x='year:O',
    y='net_generation:Q',
    color='source:N'
)

chart
```
